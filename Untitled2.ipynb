```python
import pandas as pd
import bs4
import requests
import openpyxl
import urllib.request
import urllib
```

```python
def mk(page):
 
  name_list = []
 
  name = []

  while page <= 5:
    data = requests.get('https://opend.data.go.th/playground/data.go.th/ckan/225ff892-3874-4072-9787-a2c30128e3f7')
    soup = bs4.BeautifulSoup(data.text)
    print(soup.find_all('div',{'class':'ant-card-body'}))
    for c in soup.find_all('div',{'class':'wtHider'}):
      name_list.append(c.find('div',{'class':'wtSpreader'}).find('tr').text)
    print(name_list) 

     
      
    name.append('รายการ : {}'.format(name_list))
    print('Complete page number: ',page)
    page += 1
    table = pd.DataFrame([name_list]).transpose()
    table.columns = ['name']
    table.set_index('name')
    table.to_excel('web_scraping.xlsx',engine='openpyxl')
    return (name)
```

```python
mk(1)
```
