```python
from urllib.request import urlopen as req
from bs4 import BeautifulSoup as Soup
```

```python
url = 'https://www.mk1642.com/th/Product/Promotion.aspx'
webopen = req(url)
page_html = webopen.read()
print(page_html) #report in term of bytes
webopen.close()
```

```python
data = Soup(page_html,'html.parser')
print(data)
```

```python
temp = data.findAll('div',{'class':'txt-price'})
print(type(temp))
```

```python
result = temp[3].text
print(result)
```

```python
province = data.findAll('div',{'class':'col-6 topic-divide text-left'})
pv = province[0].text.replace(' ','') #ลบช่องว่างหน้าด้านเชียงใหม่ ที่ไม่ใช่ space แต่ให้copy ช่องว่าง
print(pv)
print('จังหวัด : {} อุณหภูมิ: {}'.format(pv,result)) #จัด format
print(f'จังหวัด :{pv} อุณหภูมิ: {result}')
```
