```python
import os # use commandline  
import pandas as pd
from google.colab import files
import io
uploaded = files.upload()
data  = pd.read_csv(io.BytesIO(uploaded['car.csv']))
```

```python
X_train = pd.DataFrame(data[['buying', 'maint', 'doors', 'persons','lug_boot','safety']])
y_train = pd.DataFrame(data['class'])
```

```python
X_train
```

```python
from sklearn import preprocessing
X_train = X_train.apply(preprocessing.LabelEncoder().fit_transform)
```

```python
X_train
```

```python
print(X_train.info())
print(X_train.head())
```

```python
print(y_train)
```

```python
from sklearn.model_selection import train_test_split
# Creating Train and Test datasets
X_train, X_test, y_train, y_test = train_test_split(X_train,y_train, random_state = 55, test_size = 0.33)
```

```python
X_test
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import CategoricalNB
from sklearn.naive_bayes import ComplementNB
```

```python
#Create a Gaussian Classifier
#model = GaussianNB()
#clf1 = MultinomialNB()
#clf2 = BernoulliNB()
clf3 = CategoricalNB()
#clf4 = ComplementNB()
# Train the model using the training sets 
#model.fit(X_train, y_train)
#clf1.fit(X_train, y_train)
clf3.fit(X_train, y_train)
#clf4.fit(X_train, y_train)
#clf2.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
print("Train data accuracy:",accuracy_score(y_true = y_train, y_pred= clf3.predict(X_train)))
print("Test data accuracy:",accuracy_score(y_true = y_test, y_pred= clf3.predict(X_test)))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,clf3.predict(X_test)))
```

```python
print(classification_report(y_test,clf3.predict(X_test)))
```
