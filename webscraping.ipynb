```python
import pandas as pd
import bs4
import requests
import openpyxl
import songline
```

```python
def mk(page):
 
  name_list = []
  price_list = []
  name = []
  pprice = []
  while page <= 5:
    data = requests.get('https://www.mk1642.com/th/Product/Promotion.aspx')
    soup = bs4.BeautifulSoup(data.text)
    print(soup.find_all('div',{'class':'card-template cmn-t-translate-bshadow product-card-padding'}))
    for c in soup.find_all('div',{'class':'card-template cmn-t-translate-bshadow product-card-padding'}):
      name_list.append(c.find('div',{'class':'col-6 topic-divide text-left'}).find('a').text)
      
      try:
        p = c.find('div',{'class':'txt-price'}).text
        price = p
      except :
        print('ดูรายละเอียด')
      print(price)

      if p == 'call':
        price_list.append(price)
      else:
        price_list.append(str(price))
      
      name.append('รายการ : {}'.format(name_list))
      pprice.append('ราคา : {}'.format(price))
    print('Complete page number: ',page)
    page += 1
    table = pd.DataFrame([name_list,price_list]).transpose()
    table.columns = ['name','price']
    table.set_index('name')
    table.to_excel('web_scraping.xlsx',engine='openpyxl')
    return (name[0],pprice[0])
```
