```python
from sklearn.preprocessing import StandardScaler 
import pandas as pd
import math 
ss = StandardScaler() 
x = [532,502,570,592,469,488,501,490,590,510]
Sum = 0
SmaBar = sum(x)/10
for n in range (0,10):
  Sum += (x[n]-SmaBar)**2
SD = math.sqrt(Sum/10)
Str = ' '
for i in range (0,10):
  F = (x[i]-SmaBar)/SD 
  Str += str(x[i])+ ' ---> '
  print(Str + "{:.4f}".format(F))
  Str = ' '
```
