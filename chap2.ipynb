```python
graph = {'0' : ['1','3'],'1' : ['0','2'],'2': ['1','3'],'3' : ['0','2'],'4' : ['5','6'],'5' : ['4','6'],'6' : ['4','5'],'7' : ['8'],'8' : ['7']}
for i in graph: print(i, end = " ")
```

```python
visited = [] # List to keep track of visited nodes.
queue = [] #Initialize a queue
def bfs(visited, graph,node,finalnode): 
 
  visited.append(node)
  queue.append(node)


  while queue: 
    s = queue.pop(0)
    print (s, end = " ")
    if  (s == finalnode):break
    
    
    for neighbour in graph[s]: 
        if neighbour not in visited: 
          visited.append(neighbour)
          queue.append(neighbour)
        

 
# Driver Code

node = input("Enter initial node: ")
finalnode=input("Enter final node: ")
bfs(visited, graph,node,finalnode)
```
